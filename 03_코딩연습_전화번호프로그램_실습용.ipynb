# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

# 요구사항 분석 및 설계
* 요구사항 분석 > 구체적인 동작을 정의
** i/o 설계하는 것 중요
* 홍길동, 010-123-4567 -> 딕셔너리 key : value
* input을 받아야 되겠구나 str
* 프로그램 : while 
* 기능 : 1,2,3,4,5 : if분기문 , 숫자를 입력했을 때 해당 기능이 동작 
* 기능 1 : 연락처 추가 : dict 추가 = dict[key] = value
* 기능 2 : 전체보기: for문 돌려서 출력
* 기능 3 : 조회 : 이름을 입력했을 때 전화번호 출력, dict[key]
* 기능 4 : 수정 : dict[key] = value
* 기능 5 : 삭제 : del
* 기능 6 : 종료 : break

```python
print('전화번호 프로그램')
print('1. 추가,2. 전체보기,3. 검색, 4. 수정, 5. 삭제, 6. 종료')
print('='*50)

contact = {}
while True:
    try:
        user = int(input('기능을 선택하세요(1~6 입력): '))
        if user == 1:
            new_name = input('이름:')
            new_tel = input('전화번호:')
            contact.setdefault(new_name,new_tel)
        elif user ==2:
            for name, tel in contact.items():
                print(name,':',tel)
        elif user ==3:
            search_name = input('검색 이름:')
            print(contact.get(search_name,'없는 연락처입니다.'))
        #인덱스로 조회할 경우 없으면 오류가 남 > 메서드로 처리하는 것이 좋음
        elif user ==4:
            mod_name = input('수정 이름:')
        #만약 연락처 저장소에 수정 이름이 존재한다면 저장소에서 수정
            if mod_name in contact:
                mod_tel = input('새 전화번호:')
                contact[mod_name] = mod_tel
            else : 
                print('등록되지 않은 이름입니다')
        elif user ==5:
            del_name = input('삭제 이름:')
        #만약 연락처 저장소에 삭제 이름이 존재한다면 저장소에서 삭제
            if del_name in contact:
                del contact[del_name]
            else : 
                print('등록되지 않은 이름입니다.')
        elif user ==6:
            break
        else:
            print('잘못된 선택입니다')
    except :
        print('잘못된 선택입니다.')
```

```python
print('전화번호 프로그램')

contact = {}
def add_contact():
    name = input("이름: ")
    number = input("전화번호: ")
    contact[name] = number
    print(f"{name}의 전화번호가 추가되었습니다.\n")

def view_all_contacts():
    if not contact:
        print("연락처가 비어있습니다.\n")
    else:
        for name, number in contact.items():
            print(f"{name}: {number}")

def search_contact():
    name = input("검색할 이름 입력: ")
    number = contact.get(name)
    if number:
        print(f"{name}의 전화번호: {number}\n")
    else:
        print(f"{name}을(를) 찾을 수 없습니다.\n")

def modify_contact():
    name = input("수정할 연락처의 이름을 입력하세요: ")
    if name in contact:
        new_number = input("새로운 전화번호를 입력하세요: ")
        contact[name] = new_number
        print(f"{name}의 전화번호가 수정되었습니다.\n")
    else:
        print(f"{name}을(를) 찾을 수 없습니다.\n")

def delete_contact():
    name = input("삭제할 연락처의 이름을 입력하세요: ")
    if name in contact:
        del contact[name]
        print(f"{name}의 연락처가 삭제되었습니다.\n")
    else:
        print(f"{name}을(를) 찾을 수 없습니다.\n")

while True:
    print('1. 추가,2. 전체보기,3. 검색, 4. 수정, 5. 삭제, 6. 종료')
    
    choice = input("원하는 기능을 선택하세요 (1-6): ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_all_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        modify_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택하세요.\n")
```
